```python
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.plot(example['view_count'], label=str(year))
    year+=1

plt.legend()
```

```python
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.scatter(example['published'], example['comment_count'], label=str(year))
    year+=1

plt.legend()
```

```python
plt.figure(figsize=(8,8))
desfase = 1
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.plot(example['like_count'], label=str(year))
    year+=1
plt.legend()
```

```python
#Correlacion entre duracion y visitas
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.scatter(example['duration_num'], example['view_count'], label=str(year))
    year+=1
plt.legend()
```

```python
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    
    example['like_view_prop'] = example['comment_count'] / example['view_count']
    
    plt.scatter(example['published'], example['like_view_prop'], label=str(year))
    year+=1
plt.legend()
```

```python
suma = 0

for i in onlyfiles:
    example = pd.read_csv(mypath+i)
    s2 = example['comment_count'].sum()
    suma += s2
    print(s2)
print("The total comment count is:",suma)
```

```python



plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    
    example['like_view_prop'] = example['like_count'] / example['view_count']
    example['days_since'] = pd.to_datetime(example['published']).rsub(pd.Timestamp('now', tz='UTC').floor('d')).dt.days
    
    plt.plot(example['view_count'] / (example['days_since']), label=str(year))
    year+=1
plt.legend()
```

```python

plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    
    example['like_view_prop'] = example['like_count'] / example['view_count']
    example['days_since'] = pd.to_datetime(example['published']).rsub(pd.Timestamp('now', tz='UTC').floor('d')).dt.days
    example['popularity'] = example['sentiment_score'] * (example['view_count'] + example['like_count'] + example['comment_count']) / example['days_since']
    plt.plot(example['popularity'], label=str(year))
    year+=1
plt.legend()
```
