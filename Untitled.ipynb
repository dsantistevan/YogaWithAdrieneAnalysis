```python
mypath = getcwd()+"/DATASET CAPSTONE CARDIFF/Channel_Statistics/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath,f)) and f.endswith('.csv')]
```

```python
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.plot(example['view_count'], label=str(year))
    year+=1

plt.legend()
```

```python
plt.figure(figsize=(8,8))
desfase = 1
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.plot(example['like_count'], label=str(year))
    year+=1
plt.legend()
```

```python
#Correlacion entre duracion y visitas
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    plt.scatter(example['duration_num'], example['view_count'], label=str(year))
    year+=1
plt.legend()
```

```python
plt.figure(figsize=(8,8))
desfase = 0
year = 2015 + desfase
for i in onlyfiles[desfase:]:
    example = pd.read_csv(mypath+i)
    
    example['like_view_prop'] = example['like_count'] / example['view_count']
    
    plt.scatter(example['published'], example['like_view_prop'], label=str(year))
    year+=1
plt.legend()
```
