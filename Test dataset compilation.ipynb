```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from os import listdir, getcwd
from os.path import isfile, join
import math
```

```python
mypath = getcwd()+"/DATASET CAPSTONE CARDIFF"
L = []
for i in range(2014,2022):
    path2=mypath + "/Comments " + str(i+1)
    onlyfiles = [f for f in listdir(path2) if isfile(join(path2,f)) and f.endswith('.csv')]
    L.append((onlyfiles, path2))
```

```python
L
```

```python
df = pd.read_csv(L[0][1]+'/'+L[0][0][30])
```

```python
final_df = pd.DataFrame(columns = ['text'])
```

```python
a = 0
b = 2015
for file_list, path in L:
    for file in file_list:
        
        df = pd.read_csv(path+'/'+file)
        d = df[['comment','calificacion']].where(df['calificacion']>-0.9).where(df['calificacion']<0).dropna().sample()
        final_df = pd.concat([final_df,d], ignore_index=True)
        
        c = df[['comment','calificacion']].where(df['calificacion']<-0.9).dropna().sample(3)

        final_df = pd.concat([final_df,c], ignore_index=True)
        c = df[['comment','calificacion']].where(df['calificacion']>0).dropna().sample(6)

        final_df = pd.concat([final_df,c], ignore_index=True)
        a+=1
        print(a)
    print(b)
    b+=1
```

```python
final_df.to_excel('comments_training_dataset.xlsx')
```
