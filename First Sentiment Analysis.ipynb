```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from os import listdir, getcwd, rename
from os.path import isfile, join
import math
```

```python
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
```

```python
mypath = getcwd()+"/DATASET CAPSTONE CARDIFF"
L = []
for i in range(2014,2022):
    path2=mypath + "/Comments " + str(i+1)
    onlyfiles = [f for f in listdir(path2) if isfile(join(path2,f)) and f.endswith('.csv')]
    L.append((onlyfiles, path2))
```

```python
from transformers import pipeline
```

```python
pp = pipeline("sentiment-analysis")
```

```python
pp("Noooo.")
```

```python
def calificar(row):
    i = str(row['comment'])
    if len(tokenizer(i)['input_ids']) < 512:
        val = pp(i)
        return (1 if  val[0]['label']=='POSITIVE' else -1) * val[0]['score']
    else:
        a = re.split("\.|\n",i)
        b = 0
        sent = 0
        for j in a:
            if len(j) > 0:
                val = pp(j)
                sent += len(j) * (1 if  val[0]['label']=='POSITIVE' else -1) * val[0]['score']
                b += len(j)
        #l = 'POSITIVE' if sent > 0 else 'NEGATIVE'
        return sent / b
```

```python
a = 0
b = 2015
for file_list, path in L:
    for file in file_list:
        with open(path+'/'+file, "rb") as f:
            df = pd.read_excel(f, index_col='Unnamed: 0')
            df['calificacion'] = df.apply(calificar, axis=1)
            df.to_csv(path+'/'+file.replace('.xlsx','.csv'))
        a+=1
        print(a)
    print(b)
    b+=1
```
